```python
# Функция для замены названий месяцев
def replace_months(date_str):
    """
    Заменяет названия месяцев в строке даты на их числовые эквиваленты.

    Параметры:
    date_str (str): Строка, содержащая дату в формате "день месяц год".

    Возвращает:
    str: Дата в формате "день месяц(число) год".
    """
    # Проходим по всем месяцам и их числовым эквивалентам в словаре
    for month, num in months_dict.items():
        # Заменяем название месяца на его числовое значение
        date_str = date_str.replace(month, num)
    # Возвращаем строку с замененными месяцами
    return date_str
```

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def convert_experience_to_months(experience):
    if pd.isna(experience) or experience == 'Не указано':
        return np.nan
    
    # Убираем лишнюю информацию после запятой
    experience = experience.split(',')[0]
    
    # Ищем годы и месяцы
    years = 0
    months = 0
    
    if 'год' in experience:
        years_part = experience.split('год')[0].strip()
        years = int(years_part.split()[-1]) if years_part.split()[-1].isdigit() else 0
    
    if 'месяц' in experience:
        months_part = experience.split('месяц')[0].strip()
        months = int(months_part.split()[-1]) if months_part.split()[-1].isdigit() else 0
    
    # Преобразуем в месяцы
    total_months = years * 12 + months
    return total_months
```

```python
# Применяем функцию к столбцу "Опыт работы"
data['experience_month'] = data['Опыт работы'].apply(convert_experience_to_months)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

```python
# Список городов-миллионников
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 
                  'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 
                  'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
```

```python
# Функция для извлечения нужных данных о переезде
def extract_info(row):
    parts = row.split(' , ')
    city = parts[0].strip()
    
    # Определяем категорию города
    if city == "Москва":
        city_category = "Москва"
    elif city == "Санкт-Петербург":
        city_category = "Санкт-Петербург"
    elif city in million_cities:
        city_category = "город-миллионник"
    else:
        city_category = "другие"
    
    # Готовность к переезду
    relocation_status = parts[1].strip().lower()
    if any(keyword in relocation_status for keyword in ["готов к переезду", "готова к переезду", "хочу переехать"]):
        relocation_ready = True
    else:
        relocation_ready = False
    
    # Готовность к командировкам
    travel_ready = False  # По умолчанию считаем, что не готовы
    if len(parts) > 2:  # Проверяем, есть ли информация о командировках
        travel_status = parts[2].strip().lower()
        if any(keyword in travel_status for keyword in ["готов к командировкам", "готова к командировкам", "готов к редким командировкам"]):
            travel_ready = True
    
    return pd.Series([city_category, relocation_ready, travel_ready])
```

```python
# Применяем функцию к DataFrame
data[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = data['Город, переезд, командировки'].apply(extract_info)
```

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

```python
# Создаем список всех уникальных категорий
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 
                         'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 
                       'удаленная работа', 'вахтовый метод']
```

```python
def create_dummy_features(row, categories):
    return pd.Series({category: category in row for category in categories})
```

```python
# Создаем признаки для занятости
employment_dummies = data['Занятость'].apply(lambda x: create_dummy_features(x.split(', '), employment_categories))
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Преобразование возраста в логарифмический масштаб
log_age = np.log(data_cleaned['возраст'].dropna())  # Удаляем NaN значения перед логарифмированием
```

```python
# Вычисление z-отклонений
mean_log_age = log_age.mean()
std_log_age = log_age.std()
```

```python
# Создание графика распределения
plt.figure(figsize=(10, 6))
histplot = sns.histplot(log_age, bins=30, kde=True)
```

```python
# Определение границ для выбросов (4 сигмы в правую сторону)
upper_bound = mean_log_age + 4 * std_log_age
```

```python
# Нахождение выбросов
outliers = data_cleaned[data_cleaned['возраст'] > np.exp(upper_bound)]
```

```python
# Удаление выбросов из данных
cleaned_data = data[data['возраст'] <= np.exp(upper_bound)]
```

```python
# Создание графика распределения
plt.figure(figsize=(12, 8))
histplot = sns.histplot(log_age, bins=30, kde=True, color='skyblue', edgecolor='black')

# Добавление линий на график с метками
plt.axvline(mean_log_age, color='k', lw=2, label='Среднее', linestyle='-')
plt.axvline(upper_bound, color='orange', linestyle='--', lw=2, label='Верхняя граница (4 сигмы)')

# Подсветка выбросов
for age in outliers['возраст']:
    plt.scatter(np.log(age), 0, color='red', s=100, zorder=5)  # Красные точки для выбросов

# Аннотации для выбросов
if not outliers.empty:
    plt.text(mean_log_age + 0.5, 5, 'Логарифмическое распределение асимметрично вправо', fontsize=12, color='blue')
    plt.annotate(f'Выбросы: {outliers["возраст"].values}', 
                 xy=(np.log(outliers['возраст'].mean()), 5), 
                 xytext=(np.log(outliers['возраст'].mean()) + 0.5, 7),
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 fontsize=10)
# Настройка графика
plt.title('Распределение возраста в логарифмическом масштабе', fontsize=16)
plt.xlabel('Логарифм возраста', fontsize=14)
plt.ylabel('Частота', fontsize=14)
plt.legend(fontsize=12)  # Добавление легенды
plt.grid()

# Показать график
plt.show()
```

```python
# Оценка возрастных категорий выбросов
if not outliers.empty:
    min_outlier_age = outliers['возраст'].min()
    max_outlier_age = outliers['возраст'].max()
    print(f"Соискатели с возрастом от {min_outlier_age} до {max_outlier_age} попадают под категорию выбросов.")
else:
    print("Нет выбросов в данных.")
```
